### 1. Team Report

- [ ] Top 5 throwers for each throws by count, efficiency
- [ ] Top 5 receivers for each throws by count, efficiency

```python
client = load_supabase_client()
```

```python
distinct_team_ids = client.rpc('get_distinct_keys', {'table_name': TEAM_GAME_STATS_TABLE_NAME, 
                                                     'key_column': 'team_ext_id'}).execute()
```

```python
distinct_team_ids.data
```

```python
TEAM_EXT_ID = 'empire'
resp_team = client.table(TEAM_GAME_STATS_TABLE_NAME).select("*").eq('team_ext_id', TEAM_EXT_ID).execute().data
```

```python
df_team = pd.DataFrame(resp_team)
```

```python
df_team
```

```python
resp_team_distribution = client.table(THROWS_DISTRIBUTION_TABLE_NAME).select("*").eq('team_ext_id', TEAM_EXT_ID).execute()
df_team_distribution = pd.DataFrame(resp_team_distribution.data)
```

```python
df_team_distribution.head()
```

```python
# --- most popular throwers for each throws + by perc efficiency // favorite throws
THROW_TYPE = 'dump'
df_team_distribution[df_team_distribution['throw_type'] == THROW_TYPE].groupby(['thrower_ext_id']).size().reset_index(name='counts').sort_values(by='counts')
```

```python
# --- most popular receivers for each throws
THROW_TYPE = 'dump'
df_team_distribution[df_team_distribution['throw_type'] == THROW_TYPE].groupby(['receiver_ext_id']).size().reset_index(name='counts').sort_values(by='counts')
```

### 2. Player Profile
